```python
one_way = []                                                                          #二維轉成一維
one_way = Initial.flatten()
one_way
```

```python
df = []                                                                            #將一維度資料變成 X的五個feature 和一個Y值
ds = []
for i in range(6,2208):
    df.append(one_way[i-6:i])                                               
    ds.append(one_way[i]) 
    
X = pd.DataFrame(df)
Y = pd.DataFrame(ds)
X.columns = ['A','B','C','D','E','F']
#Y.columns = ['Y']
X = X.fillna(0)
Y = Y.fillna(0)
X
```

```python
np.where(np.isnan(X))
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2) # 30% for testing, 70% for training
X_train = X[:1546]
X_test  = X[1546:]
y_train =Y[:1546]
y_test  =Y[1546:]
```

```python
from sklearn.preprocessing import StandardScaler                                                          #標準化
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn import datasets, linear_model

regr = linear_model.LinearRegression()

# 訓練模型
regr.fit(X_train_std, y_train)
print("MSE: %.2f" % np.mean((regr.predict(X_test_std) - y_test) ** 2))
print("Test R Square_test:",regr.score(X_test_std, y_test))
print("Train R Square_train:",regr.score(X_train_std, y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier                                                     # 隨機森林
RFC = RandomForestClassifier(random_state=2,n_estimators=10,min_samples_split=20,oob_score=True)
RFC.fit(X_train, y_train)
#print("MSE: %.2f" % np.mean((RFC.predict(X_test) - y_test) ** 2))
print("R Square_train:",RFC.score(X_train_std, y_train))
print("R Square_test:",RFC.score(X_test, y_test))
```

```python
X_rnn = np.array(X_train).reshape(len(X_train), 6,1)
y_rnn = np.array(y_train).reshape(len(y_train),1)
print(X_rnn.shape)
print(y_rnn.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.recurrent import SimpleRNN, LSTM
```

```python
model = Sequential()
model.add(SimpleRNN(units=32, input_shape=(6,1)))
model.add(Dense(256))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_rnn, y_rnn, epochs=1000, batch_size=5, verbose=2)
```

```python
y_pred_rnn = model.predict(X_rnn).reshape(-1)
y_pred_rnn.shape
```
