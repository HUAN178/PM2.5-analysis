```python
X_train = X[:64]
X_test  = X[64:]
y_train =y[:64]
y_test  =y[64:]
X_train
```

```python
from sklearn.preprocessing import StandardScaler                                                          #標準化
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn import datasets, linear_model

regr = linear_model.LinearRegression()

# 訓練模型
regr.fit(X_train_std, y_train)
#print("MSE: %.2f" % np.mean((regr.predict(X_test_std) - y_test) ** 2))
print("Test R Square_test:",regr.score(X_test_std, y_test))
print("Train R Square_train:",regr.score(X_train_std, y_train))
```

```python
X_rnn = np.array(X).reshape(len(X), 6,1)                              #使用RNN
y_rnn = np.array(y).reshape(len(y),1)
print(X_rnn.shape)
print(y_rnn.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers.recurrent import SimpleRNN, LSTM
```

```python
model = Sequential()
model.add(SimpleRNN(units=20, input_shape=(6,1)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
model.summary()
```

```python
model.fit(X_rnn, y_rnn, epochs=1000, batch_size=5, verbose=2)
```

```python
y_pred_rnn = model.predict(X_rnn).reshape(-1)
```
